<h2  style=background:#ffff00;border-style:solid>   <center>Blob Visualization</center></h2>

<h3  style=background:#ffff00;border-style:solid>   <center> Imports</center></h3>

```python
import sklearn.datasets as sk_datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA,KernelPCA,IncrementalPCA,MiniBatchSparsePCA,SparsePCA
from sklearn.manifold import TSNE
```

```python
import umap
```

<h3  style=background:#ffff00;border-style:solid>   <center> Blob generation</center></h3>

```python
X1,Y1=sk_datasets.make_blobs(n_samples=1000,n_features=20,centers=4)
```

```python
Y1
```

```python
X1
```

```python
plt.scatter(X1[:, 1], X1[:, 2], marker='o', c=Y1)
```

<h3  style=background:#ffff00;border-style:solid>   <center>PCA Visualization</center></h3>

```python
pca_model=PCA(n_components=2)
pca_model.fit(X1,Y1)
```

```python
X1_pca=pca_model.transform(X1)
```

```python
plt.scatter(X1_pca[:, 0], X1_pca[:, 1], marker='o', c=Y1)
```

<h3  style=background:#ffff00;border-style:solid>   <center>t-SNE visualization</center></h3>

```python
tsne_model=TSNE(n_components=2)
tsne_model.fit(X1,Y1)
```

```python
X1_tsne=tsne_model.fit_transform(X1)
```

```python
print(X1_tsne.shape)
```

<h3  style=background:#ffff00;border-style:solid>   <center>UMAP visualization</center></h3>

```python
plt.scatter(X1_tsne[:, 0], X1_tsne[:, 1], marker='o', c=Y1)
```

```python

X1_umap = umap.UMAP(n_neighbors=5,
                      min_dist=0.3,
                      metric='correlation').fit_transform(X1)
```

```python
plt.scatter(X1_umap[:, 0], X1_umap[:, 1], marker='o', c=Y1)
```
